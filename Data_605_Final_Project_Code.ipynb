# Code to clean dataset for Tags worcloud/barplot

```python
import pandas as pd
import re

df = pd.read_csv('/Users/guldeepkaur/Downloads/CAvideos.csv' ,encoding="utf-8-sig", low_memory=False)  


title = df.tags.to_list()

def unique(list1):
    unique_titlelist = [] 
      
    for x in list1:
        if x not in unique_titlelist:
            unique_titlelist.append(x) 
    return(unique_titlelist)
 
x = unique(title)


tagwordlist = [] 

for y in x:
    wordList = y.split('|')  
    tagwordlist.append(wordList) 

flatList = [item.lower() for x in tagwordlist for item in x]

df1 = pd.DataFrame(flatList,columns=['newTagwords'])
df1 

df1.to_csv('/users/guldeepkaur/Downloads/newtagtext.csv', index=False)
```

# Code to plot triangle HeatMap

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('/Users/guldeepkaur/Downloads/CAvideos.csv' ,encoding="utf-8-sig", low_memory=False)
df=df.drop(['category_id'], axis=1)
plt.figure(figsize=(16, 10))
# define the mask to set the values in the upper triangle to True
mask = np.triu(np.ones_like(df.corr(), dtype=np.bool))

heatmap = sns.heatmap(df.corr(), mask=mask, vmin=-1, vmax=1, annot=True,annot_kws={"size":15}, cmap="mako_r")
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':20}, pad=16);



heatmap.tick_params(labelsize=15)
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=30)
```
